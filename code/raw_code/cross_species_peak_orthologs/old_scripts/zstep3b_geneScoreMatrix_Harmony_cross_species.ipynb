```r
## load the seurat object
geneScoreMergedRDS_fn = file.path(PROJDIR, 'rdas', 'geneScoreMergedSeurat.rds')
obj_seurat = readRDS(file = geneScoreMergedRDS_fn)
```

```r
names(obj_seurat@meta.data)
head(obj_seurat@meta.data)
```

# 1) compute harmony integration

```r
objHarmony <- RunHarmony(obj_seurat, group.by.vars = c('Sample'), theta = c(.5),
                         reduction.save = "harmony", reduction = "pca", dims.use = 1:30,
                         assay.use = "RNA", project.dim = FALSE,
                         max.iter.harmony = 20, max.iter.cluster = 30)
```

```r
objHarmony <- RunUMAP(
    objHarmony, dims = 1:30, n.neighbors = 100, reduction = 'harmony',
    min.dist = 0.1, spread = 2, verbose = FALSE)
objHarmony = 

obj_seurat <- RunUMAP(
    obj_seurat, dims = 1:30, n.neighbors = 100, 
    min.dist = 0.1, spread = 2, verbose = FALSE)
```

# 2) visualize harmonized clusters

```r
p_unintegrated_species = 
    DimPlot(object = obj_seurat, label = FALSE, group.by = 'Species', cols = 'Dark2') + 
    ggplot2::ggtitle("Unintegrated")

p_integrated_species = 
    DimPlot(object = objHarmony, label = FALSE, group.by = 'Species', cols = 'Dark2') +
    ggplot2::ggtitle("Seurat integration")

p_unintegrated_species + p_integrated_species
```

```r
p_unintegrated_clusters2 = 
    DimPlot(object = obj_seurat, label = TRUE, group.by = 'Clusters2', cols = 'Paired') + 
    ggplot2::ggtitle("Unintegrated")

p_integrated_clusters2 = 
    DimPlot(object = objHarmony, label = TRUE, group.by = 'Clusters2', cols = 'Paired') +
    ggplot2::ggtitle("Seurat integration")

p_unintegrated_clusters2 + p_integrated_clusters2
```

```r
p_unintegrated_sample = 
    DimPlot(object = obj_seurat, label = TRUE, group.by = 'Sample', cols = 'alphabet') + 
    ggplot2::ggtitle("Unintegrated")

p_integrated_sample = 
    DimPlot(object = objHarmony, label = TRUE, group.by = 'Sample', cols = 'alphabet') +
    ggplot2::ggtitle("Seurat integration")

p_unintegrated_sample + p_integrated_sample
```
