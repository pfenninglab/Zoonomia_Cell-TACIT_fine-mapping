# Make plots

```r
PROJDIR='../../../data/raw_data/hg38/Corces_2020'
ARCHDIR=file.path(PROJDIR,'ArchR_Corces2020_caudate_labeled2')
proj = loadArchRProject(ARCHDIR)
```

```r
# plot the confusion matrix of cell types by subjects
cM <- confusionMatrix(paste0(proj$Clusters), paste0(proj$Subject))
cM <- cM / Matrix::rowSums(cM)
p <- pheatmap::pheatmap(
  mat = as.matrix(cM), 
  color = paletteContinuous("whiteBlue"), 
  border_color = "black")
p
```

```r
# Broad MSN marker
markMSN3 = c('Ppp1r1b','Bcl11b','Kiaa1211l','Pde1b','Foxp1','Six3' )# MSN markers, Ppp1r1b = DARPP-32, Bcl11b=Ctip2
p <- plotEmbedding( ArchRProj = proj, colorBy = "GeneScoreMatrix", imputeWeights = NULL,
                    name = markMSN3, embedding = "UMAP")
p3 <- lapply(p, function(x){
  x + guides(color = FALSE, fill = FALSE) + 
    theme_ArchR(baseSize = 6.5) + theme(plot.margin = unit(c(0, 0, 0, 0), "cm")) +
    theme(  axis.text.x=element_blank(),  axis.ticks.x=element_blank(), 
            axis.text.y=element_blank(),  axis.ticks.y=element_blank())
})
do.call(cowplot::plot_grid, c(list(ncol = 3),p3))
```

# Relabel cluster and call peaks

```r
# plot the confusion matrix of cell types by subjects
cM <- confusionMatrix(paste0(proj$FinalClusters), paste0(proj$Clusters))
cM <- cM / Matrix::rowSums(cM)
p <- pheatmap::pheatmap(
  mat = as.matrix(cM), 
  color = paletteContinuous("whiteBlue"), 
  border_color = "black")
p
```

```r
############################################
# using marker gene scores, give cluster names
remapClust <- c(
    'C2' = 'Drop',
    'C3' = 'Drop',
    'C4' = 'Drop',
    'C9' = 'Drop',
    'C10'= 'Drop',
    'C11'= 'Drop',
    'C13'= 'Drop',
    'C14'= 'Drop',

    'C1' = 'MSN_SN',
    'C5' = 'MSN_D1',
    'C7' = 'MSN_D1',
    'C6' = 'MSN_D2',
    'C8' = 'INT_Pvalb',

    'C12'= 'Microglia',

    'C15'= 'Oligo',
    'C16'= 'Oligo',
    
    'C17'= 'Astro',
    
    'C18'= 'OPC'
)
proj$Clusters2 <- mapLabels(proj$Clusters, newLabels = remapClust, 
                            oldLabels = names(remapClust))
table(proj$Clusters2)
```

```r
# drop the drop clusters
ARCHDIR2=file.path(PROJDIR,'ArchR_Corces2020_caudate_labeled')
idxSample <- BiocGenerics::which(proj$Clusters2 %ni% "Drop")
cellsSample <- proj$cellNames[idxSample]

proj2 = subsetArchRProject(
  ArchRProj = proj,
  cells = cellsSample,
  outputDirectory = ARCHDIR2, force = TRUE
)
```

```r
## make UMAP plots
p1 <- plotEmbedding(ArchRProj = proj2, colorBy = "cellColData", name = "Clusters2", embedding = "UMAP")

p3 <- plotEmbedding(ArchRProj = proj2, colorBy = "cellColData", name = "Sample", embedding = "UMAP")
ggAlignPlots(p1, p3, type = "h")
```

```r
# make group coverage, call peaks, and 
proj2<-addGroupCoverages(proj2, groupBy="Clusters2", force = TRUE)
```

```r
# call peaks 
proj2<-addReproduciblePeakSet(proj2, groupBy = "Clusters2", plot = FALSE)
```

```r
# add peak counts matrix 
proj2 <- addPeakMatrix(proj2)
```

```r
# add motif enrichment matrix
proj2 <- addMotifAnnotations(ArchRProj = proj2, motifSet = "cisbp", name = "Motif")
```

```r
proj2 = saveArchRProject(ArchRProj = proj2)
```

```r
# add motif deviations matrix
proj2 <- addBgdPeaks(proj2, force = TRUE)
proj2 <- addDeviationsMatrix(proj2,  peakAnnotation = "Motif", force = TRUE)
```

```r
# add co-accessibility matrix
proj2 <- addCoAccessibility(proj2, reducedDims = "Harmony", dimsToUse = 1:30,
                           scaleDims = TRUE, corCutOff = 0.75, k = 100, 
                           knnIteration = 500, overlapCutoff = 0.8, 
                           maxDist = 1e+05, scaleTo = 10^4, log2Norm = TRUE)
```

```r
proj2 = saveArchRProject(ArchRProj = proj2)
```
