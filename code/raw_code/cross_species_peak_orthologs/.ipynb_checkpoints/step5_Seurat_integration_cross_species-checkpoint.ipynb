# 2) integration with Seurat Intergration

```r
# split seurat object up by Species
speciesList_seurat = SplitObject(obj_seurat, split.by = "Sample")
names(speciesList_seurat)
```

```r
# compute TF-IDF and find top variable features
speciesList_seurat <- lapply(X = speciesList_seurat, FUN = function(x) {
    x <- RunTFIDF(x, verbose = FALSE)
    x <- FindTopFeatures(x, min.cutoff = 'q5')
})
```

```r
lapply(speciesList_seurat, nrow)
```

```r
# get the features
n_features = 20000

## save the seurat object
featureRDS_fn = file.path(PROJDIR, 'rdas', paste0('seurat_integration_features_',n_features,'.rds'))

if(!file.exists(featureRDS_fn)){
    features <- SelectIntegrationFeatures(object.list = speciesList_seurat)
    saveRDS(features, file = featureRDS_fn)
} else {
    features = readRDS(featureRDS_fn)
}
```

```r
# compute SVD on just the top features for reciprocal PCA integration
speciesList_seurat <- lapply(X = speciesList_seurat, FUN = function(x) {
    x <- RunSVD(x, features = features, verbose = FALSE)
})
```

```r
# find integration anchors between species
anchors <- FindIntegrationAnchors(
    object.list = speciesList_seurat, reduction = 'rpca', anchor.features = features,
    # one sample from each species as reference
    reference = c(1, 4, 5), k.filter = NA
)
```

```r
# integrate data and create a new merged object
integrated <- IntegrateData(
    anchorset = anchors, dims = 2:30, preserve.order = TRUE
)
```

```r
# we now have a "corrected" TF-IDF matrix, and can run LSI again on this corrected matrix
integrated <- RunSVD( object = integrated, n = 30)
integrated <- RunUMAP( object = integrated, dims = 2:30)
```
